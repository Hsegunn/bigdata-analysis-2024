````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
## 단일 차트
plt.plot([1,2,3,4,5])
plt.show() # [<matplotlib.lines.Line2D at 0x12c14460fe0>] 객체 정보 출력이 사라짐
````

````python
plt.plot(np.random.randn(50)) # np.random.randn(n) n개의 정수를 랜덤하게 만들어줌
plt.show()
````

````python
## 다중차트, figure() 4개 까지 가능
fig = plt.figure()
plt1 = fig.add_subplot(2,2,1) # 1번 차트 
plt2 = fig.add_subplot(2,2,2) # 2번 차트 
plt3 = fig.add_subplot(2,2,3) # 3번 차트 
plt4 = fig.add_subplot(2,2,4) # 4번 차트 

plt1.plot(np.random.randn(50))
plt2.hist(np.random.randn(100))
plt3.scatter(np.arange(30), np.random.randn(30))
plt4.plot(np.random.randn(30), 'violet') # 색 'r', 'g', 'b', 'y', 'violet'

plt.show()
````

````python
## 다중차트 2번째, 4개 이상가능
# subplot() figsize로 차트 넓이 조정가능
fig, axes = plt.subplots(2,3, figsize=(18,6))
axes[0][0].plot(np.random.rand(50))
axes[0][1].hist(np.random.rand(100))
axes[1][0].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(30), 'violet')

plt.show()
````

#### 차트 스타일 지정 옵션
- label, style, alpha, rot, xticks, yticks, xlim, color, marker, ...

````python
fig, axes = plt.subplots(2,3, figsize=(18,6))
# color = 'k'검정색, 'crimson' 핏빛, alpha=0.0 완전투명, 1.0 투명도 없음
axes[0][0].plot(np.random.rand(50), color='crimson', alpha=1.0)
# bins는 그루핑 갯수
axes[0][1].hist(np.random.rand(100),color='g', bins=50)
# linestyle은 선 종류 '-' , '--' , '-.' , ':'/ 'soild', 'dotted' , 'dashed', 'dashdot'
axes[0][2].plot(np.random.rand(50), color='b', linestyle='--')
axes[1][0].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(30), 'violet')

plt.show()
````
